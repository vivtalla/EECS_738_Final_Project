```python
df_final = df_final[df_final.season > 2000]
df_final.head()
df_newFinal = pd.read_csv('../Data/data-full-good.csv')
df_newFinal.head(10)
```

```python
#use label encoder to enumerate the type of species
from sklearn.preprocessing import LabelEncoder
#encoding all other features for classification purposes
le = LabelEncoder()
le.fit(df_newFinal['winner'])
df_newFinal['winner'] = le.transform(df_newFinal['winner'])
df_newFinal.head(10)
#Home = 1
#Away = 0
```

```python
normalized_data = df_newFinal[['away_defensive_rating', 'home_block_percentage']].apply(lambda z: (z - z.min()) / (z.max() - z.min()))
normalized_data['winner'] = df_newFinal['winner']
normalized_data.tail(5)
```

```python
normalized_data.describe()
```

```python
#splitting into train and test datasets using numpy's randn
msk = np.random.rand(len(normalized_data)) < 0.8
train = normalized_data[msk]
test = normalized_data[~msk]
print('Size of train data:', len(train))
print('Size of test data:', len(test))
```

```python
X = train.values[:,:2]
X[:3]
targets = [[1,0],[0,1]]
y = np.array([targets[int(x)] for x in train.values[:,2:3]])
y[:3]
print(X)
```

```python
#the neural network inputs and the length and width of the species
nninputs = len(X[0])
hidden_neurons = 2
np.random.seed(4)
weight_matrix = 2*np.random.random((nninputs, hidden_neurons)) - 1
weight_matrix
```

```python
#three output values, one for each species type
num_outputs = len(y[0])
weight_matrix2 = 2*np.random.random((hidden_neurons, num_outputs)) - 1
weight_matrix2
```

```python
# sigmoid function
_x = np.linspace( -5, 5, 50 )
_y = 1 / ( 1 + np.exp( -_x ) )
```

```python
nn_p_rate = 0.2
for epoch in range(50000):
    #non-linear activation function, sigmoid
    sigmoid1 = 1/(1 + np.exp(-(np.dot(X, weight_matrix)))) 
    sigmoid2 = 1/(1 + np.exp(-(np.dot(sigmoid1, weight_matrix2))))
    error_rate = (abs(y - sigmoid2)).mean()
    sigmoid2_delta = (y - sigmoid2)*(sigmoid2 * (1-sigmoid2))
    sigmoid1_delta = sigmoid2_delta.dot(weight_matrix2.T) * (sigmoid1 * (1-sigmoid1))
    weight_matrix2 += sigmoid1.T.dot(sigmoid2_delta) * nn_p_rate
    weight_matrix += X.T.dot(sigmoid1_delta) * nn_p_rate
print('Neural net error rate:', error_rate)
```

```python
X = test.values[:,:2]
y = np.array([targets[int(x)] for x in test.values[:,2:3]])

l1 = 1/(1 + np.exp(-(np.dot(X, weight_matrix))))
l2 = 1/(1 + np.exp(-(np.dot(l1, weight_matrix2))))

np.round(l2,3)
```

```python
prediction = np.argmax(l2, axis=1) # prediction
res = prediction == np.argmax(y, axis=1)
correct = np.sum(res)/len(res)
#translating back from label encoded values
results = test[['winner']].replace([0,1], ['Away','Home'])

results['Prediction'] = prediction
results['Prediction'] = results['Prediction'].replace([0,1], ['Away','Home'])
print(results)
```
